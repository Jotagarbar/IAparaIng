```python
df_clean = df.copy()
```

```python
columns_to_drop = []
```

```python
for col in df_clean.columns:
  if df_clean[col].isnull().sum() > len(df_clean) * 0.5:
    columns_to_drop.append(col)
  elif df_clean[col].dtype == int:
    columns_to_drop.append(col)
```

```python
columns_to_drop
```

```python
target = 'feature_1'
```

```python
features = [col for col in df_clean.columns if col not in [target]]
```

```python
features
```

```python
x = df_clean[features]
y = df_clean[target]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sencillo = LinearRegression()
```

```python
sencillo.fit(X_train, y_train)
```

```python
sencillo.coef_
```

```python
y_pred_test = sencillo.predict(X_test)
```

```python
sencillo.intercept_
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
```

```python
r2 = r2_score(y_test, y_pred_test)
mae = mean_absolute_error(y_test, y_pred_test)
rmse = root_mean_squared_error(y_test, y_pred_test)
```
